# 2. Conduct Data Cleansing:

### 2.1: Drop the Class column from the created Business Class passenger satisfaction survey (“df_Business”).

```python
df_Business = df_Business.drop('Class', axis=1)
```

```python
df_Business
```

```python
df_Business = df_Business.reset_index(drop=True)
df_Business.shape
```

### 2.2: Rename “df_Business” columns by making all the columns’ names in lowercase, and if the column name has whitespace, replace it with an underscore. (For example: “Customer Type” will be renamed to “customer_type”).

```python
df_Business.rename(columns = lambda x: x.lower().replace(' ','_'), inplace= True)
df_Business.columns
```

```python
print("Upated Column Names: \n")
for column in df_Business.columns:
    print(column)
```

### 2.3: Modify improperly formatted data and handle bad data.

```python
col = df_Business.columns

for col in df_Business.columns:
    print(col,'column has', len(df_Business[col].unique()),'unique values')
    print(df_Business[col].unique())
    print('Frequency distribution of', col, 'is as follows:')
    print(df_Business[col].value_counts())
    print('-------------------------------------------------')
```

#### inflight_wifi_service, departure/arrival_time_convenient, cleanliness. These columns have no improperly formatted data, hence have not been considered in below code. 

```python
condition = (df_Business['age'] >=180)
df_Business = df_Business.drop(df_Business[condition].index)
```

```python
condition2 = (df_Business['age'] == 0)
df_Business = df_Business.drop(df_Business[condition2].index)
```

```python
mode_value = df_Business['ease_of_online_booking'].dropna().mode()[0]
#mode_value = df_Business['ease_of_online_booking'].mode()[0]
mode_value
df_Business['ease_of_online_booking']=df_Business['ease_of_online_booking'].replace(['Unknown'],[mode_value])
```

```python
df_Business['food_and_drink'] = pd.to_numeric(df_Business['food_and_drink'], errors='coerce')
mean_value = round(df_Business['food_and_drink'].mean())
mean_value_str = str(mean_value)
df_Business['food_and_drink']=df_Business['food_and_drink'].replace(['Great'],[mean_value_str])
```

```python
df_Business['gender']= df_Business['gender'].replace(['F','Fe male','M'],['Female','Female','Male'])
df_Business['customer_type']= df_Business['customer_type'].replace(['Loyal','disloyal'],['Loyal Customer','Disloyal Customer'])
df_Business['type_of_travel']=df_Business['type_of_travel'].replace(['Business','Business travel'],['Business Travel','Business Travel'])
df_Business['inflight_entertainment']=df_Business['inflight_entertainment'].replace([10.],[5.])
df_Business['baggage_handling']=df_Business['baggage_handling'].replace(['A'],['5'])
df_Business['arrival_delay_in_minutes']=df_Business['arrival_delay_in_minutes'].replace(['Three'],['3'])
df_Business['satisfaction'] = df_Business['satisfaction'].replace(['neutral','dissatisfied','satisfied','0','1'],['Neutral','Dissatisfied','Satisfied','Dissatisfied','Satisfied'])
```

### 2.4 Handle the missing values (You should have zero missing values in your cleaned dataset). rating mode, integer float mean

```python
df_Business['gender'].fillna('Unknown',inplace=True)
df_Business['customer_type'].fillna('Unknown',inplace=True)
df_Business['type_of_travel'].fillna('Unknown',inplace=True)
df_Business['satisfaction'].fillna('Unknown',inplace=True)
mean_value = round(df_Business['age'].mean())
mean_value
df_Business['age'].fillna(mean_value,inplace=True)
df_Business['inflight_wifi_service'].fillna(df_Business['inflight_wifi_service'].mode()[0],inplace=True)
df_Business['departure/arrival_time_convenient'].fillna(df_Business['departure/arrival_time_convenient'].mode()[0],inplace=True)
df_Business['ease_of_online_booking'].fillna(df_Business['ease_of_online_booking'].mode()[0],inplace=True)
df_Business['food_and_drink'].fillna(df_Business['food_and_drink'].mode()[0],inplace=True)
df_Business['inflight_entertainment'].fillna(df_Business['inflight_entertainment'].mode()[0],inplace=True)
df_Business['baggage_handling'].fillna(df_Business['baggage_handling'].mode()[0],inplace=True)
df_Business['cleanliness'].fillna(df_Business['cleanliness'].mode()[0],inplace=True)
df_Business['departure_delay_in_minutes'].fillna(df_Business['departure_delay_in_minutes'].mode()[0],inplace=True)
df_Business['flight_distance'].fillna(df_Business['flight_distance'].mode()[0],inplace=True)
df_Business['arrival_delay_in_minutes'].fillna(df_Business['arrival_delay_in_minutes'].mode()[0],inplace=True)
```

```python
# In the flight_distance column, there are total 8 rows which have values for the distance of flight 
# as 0, -5421, 2820000, 9000000. So we have dropped these rows using 2 conditions.


condition1 = (df_Business['flight_distance'] > 100000)
df_Business = df_Business.drop(df_Business[condition1].index)
```

```python
condition2 = (df_Business['flight_distance'] <= 0)
df_Business = df_Business.drop(df_Business[condition2].index)
```

```python
condition3 = (df_Business['departure_delay_in_minutes'] < 0)
df_Business = df_Business.drop(df_Business[condition3].index)
```

```python
condition4 = (df_Business['arrival_delay_in_minutes'] == '-45')
df_Business = df_Business.drop(df_Business[condition4].index)
```

```python
condition5 = (df_Business['arrival_delay_in_minutes'] == '9..')
df_Business = df_Business.drop(df_Business[condition5].index)
```

### 2.5 Remove the fully duplicated observations.

```python
df_Business = df_Business.drop_duplicates()
```

### 2.6 Fix the columns datatypes.

```python
df_Business = df_Business.convert_dtypes()
```
